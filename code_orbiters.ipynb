```python
import matplotlib.pyplot as plt

class HealthTracker:
    def __init__(self):
        self.steps = []
        self.calories_burned = []
        self.weights = []
        self.sleep_hours = []

    def record_data(self, day, steps, calories_burned, weight, sleep_hours):
        self.steps.append(steps)
        self.calories_burned.append(calories_burned)
        self.weights.append(weight)
        self.sleep_hours.append(sleep_hours)

    def plot_health_records(self):
        days = range(1, len(self.steps) + 1)

        plt.figure(figsize=(12, 8))

        plt.subplot(4, 1, 1)
        plt.plot(days, self.steps, marker='o', color='b')
        plt.title('Steps Throughout the Week')
        plt.xlabel('Days')
        plt.ylabel('Steps')
        plt.grid(True)

        plt.subplot(4, 1, 2)
        plt.plot(days, self.calories_burned, marker='o', color='r')
        plt.title('Calories Burned Throughout the Week')
        plt.xlabel('Days')
        plt.ylabel('Calories Burned')
        plt.grid(True)

        plt.subplot(4, 1, 3)
        plt.plot(days, self.weights, marker='o', color='g')
        plt.title('Weight Throughout the Week')
        plt.xlabel('Days')
        plt.ylabel('Weight')
        plt.grid(True)

        plt.subplot(4, 1, 4)
        plt.plot(days, self.sleep_hours, marker='o', color='purple')
        plt.title('Sleep Hours Throughout the Week')
        plt.xlabel('Days')
        plt.ylabel('Sleep Hours')
        plt.grid(True)

        plt.tight_layout()
        plt.show()

def main():
    tracker = HealthTracker()

    for day in range(1, 8):
        print(f"Day {day}")
        steps = int(input("Enter steps count: "))
        calories_burned = int(input("Enter calories burned: "))
        weight = float(input("Enter weight (kg): "))
        sleep_hours = float(input("Enter sleep hours: "))
        tracker.record_data(day, steps, calories_burned, weight, sleep_hours)

    tracker.plot_health_records()

if __name__ == "__main__":
    main()
```
